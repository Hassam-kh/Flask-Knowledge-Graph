```python
onto = owlready2.get_ontology("C:\\Users\\dortemon\\Downloads\\krr_proj\\KRR-KGLD-Fall2022-KK22-07-Ontology.owl").load()
```

```python
uni_designation = members_of_ruling_body["Designation"].unique()
uni_assembly = members_of_constitution_assembly["Assembly"].unique()
uni_state = MNA["State"].unique()
uni_office = Officers_Phone_List_Members["Office Designation"].unique()

with onto: 

    for i in uni_office:

        temp = Officers_Phone_List_Members[Officers_Phone_List_Members["Office Designation"] == i ]
        office = onto.Offices(i)
        office.has_Name = [i]
        for j in range(len(temp)):

            name = onto.Officers_Phone_List_Members(temp.iloc[j]["Name"])
            des = onto.Designation(temp.iloc[j]["Designation"])
            des.has_Name = [temp.iloc[j]["Designation"]]
            name.has_Name = [temp.iloc[j]["Name"]]
            name.works_under_Office.append(office)
            name.has_Designation.append(des)
            if temp.iloc[j]["Office"] != "-":
                name.has_Phone_no = [temp.iloc[j]["Office"]]
            else:
                name.has_Phone_no = ["none"]

            office.Designations_under_Office.append(des)

    for i in uni_designation:

        temp = members_of_ruling_body[members_of_ruling_body["Designation"] == i ]
        designation = onto.Designation(i)
        designation.has_Name = [i]
        for j in range(len(temp)):

            name = onto.Members_of_Ruling_Body(temp.iloc[j]["name"])
            name.has_Name = [temp.iloc[j]["name"]]
            name.has_Designation.append(designation)
            name.has_Name = [temp.iloc[j]["name"]]
            name.Start_of_Tenure = [temp.iloc[j]["start_of_tenure"]]
            name.End_of_Tenure = [temp.iloc[j]["end_of_tenure"]]

    for i in uni_assembly:

        temp = members_of_constitution_assembly[members_of_constitution_assembly["Assembly"] == i ]
        assembly = onto.Assembly(i)

        assembly.From = [temp["From"].iloc[0]]
        assembly.To = [temp["To"].iloc[0]]
        for j in range(len(temp)):

            name = onto.Members_of_Constitution_Assembly(temp.iloc[j]["Name"])
            name.has_Name = [temp.iloc[j]["Name"]]
            province = onto.Provinces(temp.iloc[j]["province"])
            province.has_Name = [temp.iloc[j]["province"]]
            assembly.has_Members.append(name)
            assembly.has_Provinces.append(province)

            if temp.iloc[j]["Constituency"] != "-":
                constituency = onto.Constitution(temp.iloc[j]["Constituency"])
                constituency.has_Name = [temp.iloc[j]["Constituency"]]
                province.has_Constitution.append(constituency)
                name.handles_Constituency.append(constituency)

            name.Member_of_Province.append(province)
            name.has_Address = [temp.iloc[j]["Permanent Address"]]

    for i in uni_state:

        temp = MNA[MNA["State"] == i ]
        state = onto.Provinces(i)
        state.has_Name = [i]
        for j in range(len(temp)):

            name = onto.Members_of_National_Assembly(temp.iloc[j]["Name"])
            name.has_Name = [temp.iloc[j]["Name"]]
            cons = onto.Constitution(temp.iloc[j]["Constituency"])
            cons.has_Name = [temp.iloc[j]["Constituency"]]
            party = onto.Party(temp.iloc[j]["Party"])
            party.has_Name = [temp.iloc[j]["Party"]]
            name.Member_of_Province.append(state)
            name.handles_Constituency.append(cons)
            name.Member_of_Party.append(party)

            name.has_Address = [temp.iloc[j]["Address (Permanent/Local)"]]
            name.has_Phone_no = [temp.iloc[j]["Phone"]]

            state.has_Constitution.append(cons)
            party.Party_has_Members.append(name)

    for i in onto.Party.instances(): 
        party_list.append(i.name)

    for i in dic:
        
        party = onto.Party(i)
        gen = int(dic[i][0])
        res = int(dic[i][1])
        mino = int(dic[i][2])
        tot = int(dic[i][3])
         
        party.has_Name = [i]
        party.General_Seats = [gen]
        party.Reserved_Seats = [res]
        party.Minority_Seats = [mino]
        party.Total_Seats = [tot]
```
